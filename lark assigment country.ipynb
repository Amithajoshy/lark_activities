# visualisation


```python
regData = data.groupby("Region")["Population in thousands (2017)"].sum()
plt.figure(figsize = (16, 16))
plt.pie(regData,labels = regData.index, autopct = "%1.2f%%")
plt.title('REGIONWISE POPULATION',fontweight='bold',fontsize=19)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=data.sort_values('Population in thousands (2017)',ascending=False).head(20))
plt.title('Top 20 countries based on Population in thousands (2017)',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Population in thousands (2017)',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle="--")
plt.show()

```

# ON average which region has the highest


```python
#a.GDP per capita(current US$)
```

```python

data2=data.groupby("Region")[['GDP per capita (current US$)']].mean()
data2.nlargest(1,"GDP per capita (current US$)")

```

```python
#b.international trade:imports(million US$)
```

```python
data3=data.groupby("Region")[['International trade: Imports (million US$)']].mean()
data3.nlargest(1,"International trade: Imports (million US$)")
```

# Find out which region has more consistent  with respect to GDP Per capita(current US$)

```python
data4=data.groupby("Region")[["GDP per capita (current US$)"]].std()
data4.nsmallest(1,"GDP per capita (current US$)")
```
