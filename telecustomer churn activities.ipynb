# 1. Compare churn count with respect to gender

```python
plt.figure(figsize=(5,5))
sns.countplot( x='gender',hue='Churn',data=data) 
plt.title('Churn count respect to the gender',fontweight="bold")
plt.grid(linestyle="--")  
plt.show()
```

# 2. Find out how many female senior citizens there in the dataset

```python
data1=data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
data1['gender'].count()
plt.figure(figsize=(6,6))
sns.countplot( x='gender',hue='SeniorCitizen',data=data1) 
plt.title("Count of female senior citizens ")

```

# 3. Compare 'tenure' with 'Total Charges

```python
plt.figure(figsize=(8,6))
plt.scatter(data['tenure'] , data['TotalCharges'] , s=11 , c='orange')
plt.title("Tenure VS Total charges",fontweight="bold")
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
```
