```python
#filling of missing values
```

```python
d=df['Name'].str.split(',',expand=True)
d
```

```python
d1=d[1].str.split('.',expand=True)
d1
```

```python
d2=d1.drop([1,2],axis=1)
d2
```

```python
data=pd.concat([df,d2],axis=1)
data
```

```python
data=data.rename(columns={data.columns[-1]:'title'})
data
```

```python
data.groupby('title')['Age'].mean()
```

```python
data['Age']=data['Age'].fillna(data.groupby('title')['Age'].transform('mean'))
data
```

```python
data.isna().sum()
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

imp = imp.fit(data[['Cabin']])
data['Cabin'] = imp.transform(data[['Cabin']]).ravel()
```

```python
#Using percentile and IQR method we are dealing with the outlier
```

```python
Q1 = np.percentile(data2['SibSp'],25,interpolation = 'midpoint')
Q2 = np.percentile(data2['SibSp'],50,interpolation = 'midpoint')
Q3 = np.percentile(data2['SibSp'],75,interpolation = 'midpoint')

IQR = Q3-Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

print("IQR is ",IQR)
print("Lower limit is" ,lower_limit)
print("Upper Limit is",upper_limit)
```

```python
index=data2['SibSp'][(data2['SibSp']>upper_limit)|(data2['SibSp']<lower_limit)].index
data2.drop(index,inplace=True)
```

```python
#The outliers will be dropped for all the values which are outside the upper and lower limit
```

```python
sns.boxplot(y=data2["SibSp"])
```

```python
Q1 = np.percentile(data2['Age'],25,interpolation = 'midpoint')
Q2 = np.percentile(data2['Age'],50,interpolation = 'midpoint')
Q3 = np.percentile(data2['Age'],75,interpolation = 'midpoint')

IQR = Q3-Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

print("IQR is ",IQR)
print("Lower limit is" ,lower_limit)
print("Upper Limit is",upper_limit)
```

```python
#In case of age , we replace values outside upper limit with upper limit
#Values below lower limit with lower limit
```

```python
data2['Age']=np.where(data2['Age']>upper_limit,upper_limit,np.where(data2['Age']<lower_limit,lower_limit,data2['Age']))
```

```python
sns.boxplot(y=data2["Age"])
```

```python
Q1 = np.percentile(data2['Fare'],25,interpolation = 'midpoint')
Q2 = np.percentile(data2['Fare'],50,interpolation = 'midpoint')
Q3 = np.percentile(data2['Fare'],75,interpolation = 'midpoint')

IQR = Q3-Q1
lower_limit = Q1 - (1.5*IQR)
upper_limit = Q3 + (1.5*IQR)

print("IQR is ",IQR)
print("Lower limit is" ,lower_limit)
print("Upper Limit is",upper_limit)
```

```python
#In case of fate , we replace values outside upper limit with upper limit
#Values below lower limit with lower limit
```

```python
data2['Fare']=np.where(data2['Fare']>upper_limit,upper_limit,np.where(data2['Fare']<lower_limit,lower_limit,data2['Fare']))
```

```python
sns.boxplot(y=data2["Age"])
```
